```python
import pandas as pd
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding
from tensorflow.keras.layers import LSTM
```

```python
# Hyperparams

# Note: Window size needs to be at least 4 for kurtosis to work
WINDOW_SIZE = 4
```

```python
data = pd.read_csv("daily_eur_usd.csv").iloc[::-1].reset_index(drop=True)
```

```python
data
```

```python
# Use only open prices for now
data.drop(columns=["high", "low", "close"], inplace=True)
```

```python
open_skew = data.rolling(window=WINDOW_SIZE).skew(axis=1)
open_skew
```

```python
open_kurt = data.rolling(window=WINDOW_SIZE).kurt(axis=1)
open_kurt
```
